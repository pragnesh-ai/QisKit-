```python
from qiskit import *
```

```python
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
s_number="101101"
```

```python
circuit=QuantumCircuit(len(s_number)+1,len(s_number))
```

```python
# circuit.h([0,1,2,3,4,5])
circuit.h(range(len(s_number)))
circuit.x(len(s_number))
circuit.h(len(s_number))
```

```python
circuit.draw(output='mpl')
```

```python
circuit.barrier()
for i,value in enumerate(s_number):
    if value=='1':
        circuit.cx(i,len(s_number))
# circuit.cx(5,len(s_number))
# circuit.cx(3,len(s_number))
# circuit.cx(0,6)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.h(range(len(s_number)))
circuit.measure(range(len(s_number)),range(len(s_number)))
```

```python
circuit.draw(output='mpl')
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(circuit,backend=simulator,shots=1).result()
counts=result.get_counts()
print(counts)
```
